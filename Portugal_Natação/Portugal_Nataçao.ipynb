```python
from bs4 import BeautifulSoup as bs
import requests
import csv
import re 

my_dict = {}

def find_name(soup):
  nomes = soup.find('div', class_ = 'member-info').find('h4').text
  return nomes

def find_acronym(upper):
    acronym=''
    for a in upper:
        #check uppercase characters
        if a.isupper():
            acronym += a
    return acronym

def find_address(soup):
    endereco = soup.find('span', class_='icon-info-sign')
    enderecos_rua = endereco.text.splitlines()
    rua_enderecos = enderecos_rua[0]
    return rua_enderecos  

#def find_city(soup):
#    endereco = soup.find_all('span', class_='icon-info-sign')
#    
#    for enderecos in endereco:
#      enderecos_rua = enderecos.text.splitlines()
#      if enderecos_rua[2] is not False:
#        state_city = enderecos_rua[2]
#        print(state_city)
#      elif enderecos_rua[2] is None:
#        state_city.append('')
#        continue

def find_phone(soup):
    phone = soup.find('span', class_ = 'fa fa-phone-square').text
    return phone
       

def find_email(soup):
    email = soup.find('span', class_='fa fa-envelope').text

    return email
         

def find_website(soup):
    site = soup.find('span', class_='icon-ok')
    try:
        websites = site.find('h4').text
    except:
        websites = ''
    return websites

def find_logo(info):
    logo = 'https://www.anatacaodamadeira.pt/' + info.find('div', class_='member-image').find('img')['src'] 
    return logo

my_list = []
def main():
  
  url = 'https://www.anatacaodamadeira.pt/index.php/associacao/associados/clubes'
  agent = agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
  req = requests.get(url, headers = agent)
  soup = bs(req.content, 'html5lib')
  
  info = soup.findAll('div', class_='item col-sm-6 col-md-4')
  for a in info:
      my_dict={}
      
      my_dict['Logo'] = find_logo(a)
      my_dict['Name'] = find_name(a).strip()
      my_dict['Acronym'] = find_acronym(my_dict['Name'])
      my_dict['Address'] = find_address(a).strip()
      #cidade_estado = find_city(soup)  - Não tá funcionando
      my_dict['Phone'] = find_phone(a).strip()
      my_dict['Email'] = find_email(a).strip()
      my_dict['Website'] = find_website(a).strip()
      #all_clubs = soup.find_all('div', class_ = 'style-2 team-items container')
      
      my_dict['Country'] = 'Portugal'
      my_dict['Affilated'] = 'Associação de Natação da Madeira'
      my_dict['Recorded By'] = 'Eduardo Sávio'
      my_dict['Recorded Date'] = '04/02/2022'  
      my_list.append(my_dict) 
    
  return my_list
  



res = main()
#print(res)

keys = res[0].keys()

with open('Portugal_Natacao.csv', 'w', newline = '', encoding = 'utf-8-sig') as csvfile:
  fieldnames = keys
  writer = csv.DictWriter(csvfile, fieldnames = fieldnames)

  writer.writeheader()
  writer.writerows(res)

csvfile.close()
```
