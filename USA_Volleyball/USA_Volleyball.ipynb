```python
from bs4 import BeautifulSoup as bs
import requests
import csv
from selenium import webdriver
import re
import pandas as pd

def find_logo(soup):
  logo = soup.find('img')
  return logo['src']

def find_name(soup):
  name = soup.find('p', class_ = 'title')
  return name.text

def acronym_generator(soup):
  acronym = soup[0]
  for i in range(1, len(soup)):
    if soup[i-1] == ' ':
      acronym += soup[i]
  acronym = acronym.upper()
  return acronym

def find_contact(soup):
  contact = soup.find('p', class_ = 'contact')
  return contact.text

def find_address(soup):
  try:
    address = soup.find('p', class_ = 'address')
    return address.text
  except:
    return ''
  

def find_phone(soup):
  try:
    phone = soup.find('p', class_ = 'phone')
    return phone.text.replace('Phone: ', '')
  except:
    return ''

def find_email(soup):
  email = soup.find('a', class_ = 'btn-more red')
  return email['href'].replace('mailto:', '')

def find_website(soup):
  website = soup.find('a', attrs = {'target': '_blank'})
  return website['href']

def main():
  my_list = []

  #Chrome_Driver
  userAgent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('user-agent= {}'.format(userAgent))
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  driver = webdriver.Chrome('chromedriver', options = options)

  url = 'https://usavolleyball.org/membership/'

  driver.get(url)
  selenium_get = driver.page_source
  soup = bs(selenium_get, 'html.parser')
  
  #info = soup.find_all('div', class_ = 'region')
  info = soup.find_all('div', class_ = 'region')
  
  
  for x in info:

    my_dict = {}
    my_dict['Logo'] = find_logo(x)
    my_dict['Name'] = find_name(x)
    my_dict['Acronym'] = acronym_generator(find_name(x))
    my_dict['Contact'] = find_contact(x)
    my_dict['Address'] = find_address(x)
    my_dict['Phone'] = find_phone(x)
    my_dict['Email'] = find_email(x)
    my_dict['Website'] = find_website(x)
    
    my_dict['Country'] = 'USA'
    my_dict['Affilated'] = 'USA Volleyball'
    my_dict['Sport'] = 'Volleyball, Beach Volleyball'
    my_dict['Reach'] = 'State'
    my_dict['Category'] = 'Olympic'
    my_dict['Recorded By'] = 'Eduardo Sávio'
    my_dict['Recorded Date'] = '09/03/2022' 
    
    
    my_list.append(my_dict)
  return my_list

#main()
res = main()

#save as xlsx
df = pd.DataFrame(data=res)
df.to_excel('USA_Volleyball.xlsx')
```
