```python
from bs4 import BeautifulSoup as bs
import requests
import csv
from selenium import webdriver
import re

def find_name(soup):
  try:
    name = soup.find('a')
    return name.text
  except:
    return ''

def acronym_generator(soup):
  try:
    acronym = soup[0]
    for i in range(1, len(soup)):
      if soup[i-1] == ' ':
        acronym += soup[i]
    acronym = acronym.upper()
    return acronym
  except:
    return ''

def find_website(soup):
  try:
    website = soup.find('a')
    return website['href']
  except: 
    return ''

def main():
  my_list = []
  
  url = 'https://www.usef.org/about-us/affiliate-associations'
  
  agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('user-agent= {}'.format(agent))
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  driver = webdriver.Chrome('chromedriver', options = options)

  driver.get(url)
  selenium_get = driver.page_source
  soup = bs(selenium_get, 'html.parser')
  
  info = soup.find('div', class_ = 'page-content').find_all('p')
  #print(info)
  for x in info:
    my_dict = {}
    my_dict['Name'] = find_name(x)
    my_dict['Acronym'] = acronym_generator(find_name(x))
    my_dict['Website'] = find_website(x)

    my_dict['Country'] = 'USA'
    my_dict['Affilated'] = 'USA Equestrian'
    my_dict['Sport'] = 'Equestrian Combined Driving, Equestrian Dressage, Equestrian Endurance, Equestrian Eventing, Equestrian Jumping, Para-Equestrian, Equestrian Reining, Equestrian Vaulting, Equestrian English Pleasure, Equestrian Carriage Pleasure Driving, Equestrian Hunter, Equestrian Hunter/Jumping Seat Equitation, Equestrian Parade Horse, Equestrian Roadster, Equestrian Saddle Seat, Equestrian Western, Equestrian Western Dressage, Equestrian Western/Reining Seat Equitation'
    my_dict['Reach'] = 'City'
    my_dict['Category'] = 'Olympic, Paralympic'
    my_dict['Recorded By'] = 'Eduardo Sávio'
    my_dict['Recorded Date'] = '21/03/2022' 
    my_list.append(my_dict)

  return my_list

res = main()

#save as xlsx
import pandas as pd
df = pd.DataFrame(data=res)
df.to_excel('US Equestrian.xlsx')
```
