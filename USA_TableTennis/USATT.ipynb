```python
# Row 1
from bs4 import BeautifulSoup as bs
import requests
import csv
from selenium import webdriver
import re

def find_name(soup):
  try:
    name = soup.find('div', class_ = 'col-xs-12 col-sm-7 col-md-7 col-lg-7')
    return name.text.strip()
  except:
    return ''

def acronym_generator(soup):
  acronym = soup[0]
  for i in range(1, len(soup)):
    if soup[i-1] == ' ':
      acronym += soup[i]
  acronym = acronym.upper()
  return acronym

def find_address(soup):
  try:
    address = soup.find_all('div', class_ = 'col-xs-12 col-sm-7 col-md-7 col-lg-7')
    return address[1].text.strip()
  except:
    return ''

def find_admin(soup):
  try:
    admin = soup.find_all('div', class_ = 'col-xs-12 col-sm-7 col-md-7 col-lg-7')
    return admin[2].text.strip()
  except:
    return ''

def find_email(soup):
  try:
    email = soup.find_all('div', class_ = 'col-xs-12 col-sm-7 col-md-7 col-lg-7')
    return email[3].text.strip()
  except:
    return ''

def find_phone(soup):
  try:
    phone = soup.find_all('div', class_ = 'col-xs-12 col-sm-7 col-md-7 col-lg-7')
    return phone[5].text.strip()
  except:
    return ''

def containsNumber(value):
  for character in value:
    if character.isdigit():
      return 'True'
  return 'False'

def find_website(soup):
  website = soup.find_all('div', class_ = 'col-xs-12 col-sm-7 col-md-7 col-lg-7')
  website_check = website[6].text.strip()
  try:
    if containsNumber(website_check) == 'True':
      website_check = website[7].text.strip()
    else:
      website_check = website[6].text.strip()
    return website_check
  except:
    return ''

def main():
  my_list = []
  
  url = 'https://usatt.simplycompete.com/c/d?searchBy=physical_state&query=&state=&max=100&v='
  agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('user-agent= {}'.format(agent))
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  driver = webdriver.Chrome('chromedriver', options = options)  

  
  driver.get(url)
  selenium_get = driver.page_source
  soup = bs(selenium_get, 'html.parser')

  
  info = soup.find_all('div', class_ = 'col-xs-12 col-sm-12 col-md-6 col-lg-4 card')
  regex_links = re.compile(r'\/c\/p\/[a-zA-Z0-9-?=&_]+')

  # Link para cada bloco
  for i in info:
    links = i.find('div', class_ = 'category-div-block usesSpinner')
    links = regex_links.findall(links['onclick'].strip())
    for link in links:
      my_dict = {}
      link = str(link).strip("['']")
      url = 'https://usatt.simplycompete.com{}'.format(link)
      page = requests.get(url = url)
      soup = bs(page.text, 'html.parser')
      info = soup.find('div', class_ = 'wrapper outer').find('div', class_ = 'inner-wrap').find('div', class_ = 'container')#.find('div', class_ = 'row')#.find('div', class_ = 'col-xs-12 col-sm-12 col-md-8 col-lg-8 right-panel-wrapper personal-information-panel no-pad')#.find('div', class_ = 'content').find('div', class_ = 'tab-content')
      #print(info)
      info2 = info.find('div', class_ = 'tab-pane active')
      #print(info2)
      #info2 = ''
      
      my_dict['Name'] = find_name(info2)
      my_dict['Acronym'] = acronym_generator(str(find_name(info2)))
      my_dict['Address'] = find_address(info2)
      my_dict['Admin'] = find_admin(info2)
      my_dict['Email'] = find_email(info2)
      my_dict['Phone'] = find_phone(info2)
      my_dict['Website'] = find_website(info2)

      my_dict['Country'] = 'USA'
      my_dict['Affilated'] = 'USA Table Tennis'
      my_dict['Sport'] = 'Table Tennis'
      my_dict['Reach'] = 'City'
      my_dict['Category'] = 'Olympic, Paralympic'
      my_dict['Recorded By'] = 'Eduardo Sávio'
      my_dict['Recorded Date'] = '09/03/2022' 

      #print(my_dict)

      my_list.append(my_dict)
  return my_list

res = main()

#save as xlsx
import pandas as pd
df = pd.DataFrame(data=res)
df.to_excel('USA Table Tennis.xlsx')
```
