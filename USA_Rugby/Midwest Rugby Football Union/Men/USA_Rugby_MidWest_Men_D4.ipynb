```python
from bs4 import BeautifulSoup as bs
import requests
import csv

def find_name(soup):
  name = soup.find_next('td', class_ = 's6')
  
  return name.text

def find_name_cutted(soup):
  name_cutted = soup.find_next('td', class_ = 's10')
  return name_cutted.text
def main():
  
  my_list = []
  url = 'https://docs.google.com/spreadsheets/u/0/d/e/2PACX-1vQ1RYDcajQV56ZL68EK9oWiGLY4YDoSwDKN1rgJfh-kW6DtbP1xrfuZbuALjxRKJYGXtnmRv-HuDnKp/pubhtml/sheet?headers=false&gid=162693995'
  agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
  req = requests.get(url, headers = agent)
  soup = bs(req.content, 'html5lib')
  
  info = soup.find('tbody')
  info = info.find_all('tr')
  #print(info.prettify())
  
  for a in info:
    my_dict = {}

    try:
      my_dict['Name'] = find_name(a)
    except:
      pass
    try:
      my_dict['Name_Cut'] = find_name_cutted(a)
    except:
      pass
    try:
      my_dict['Acronym'] = find_name(a)
    except:
      pass


    my_dict['Country'] = 'USA'
    my_dict['Affilated'] = 'Midwest Rugby Football Union'
    my_dict['Recorded By'] = 'Eduardo Sávio'
    my_dict['Recorded Date'] = '15/02/2022'  
    my_list.append(my_dict)
  
  return my_list

res = main()

keys = res[0].keys()

with open('USA_Rugby_MidWest_Men_D4.csv', 'w', newline = '', encoding = 'utf-8-sig') as csvfile:
  fieldnames = keys
  writer = csv.DictWriter(csvfile, fieldnames = fieldnames)

  writer.writeheader()
  writer.writerows(res)

csvfile.close()
```
