```python
# Times dentro de um tr

from bs4 import BeautifulSoup as bs
import requests
import csv

def find_name(soup):
  name = soup.find_next('td', attrs = {'data-title': 'Team'})
  return name.text

def main():
  my_list = []

  url = 'http://empire.rugby/league/160231/'
  agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
  req = requests.get(url, headers = agent)
  soup = bs(req.content, 'html5lib')

  info = soup.find_all('tr')
  
  
  for a in info:
    my_dict = {}

    my_dict['Name'] = find_name(a)
    my_dict['Acronym'] = find_name(a)

    my_dict['Country'] = 'USA'
    my_dict['Affilated'] = 'Empire Rugby Union'
    my_dict['Recorded By'] = 'Eduardo Sávio'
    my_dict['Recorded Date'] = '15/02/2022'  
    my_list.append(my_dict) 
    
  return my_list

res = main()

keys = res[0].keys()

with open('USA_Rugby_EmpireRugby_MenChampionship.csv', 'w', newline = '', encoding = 'utf-8-sig') as csvfile:
  fieldnames = keys
  writer = csv.DictWriter(csvfile, fieldnames = fieldnames)

  writer.writeheader()
  writer.writerows(res)

csvfile.close()

  
```
