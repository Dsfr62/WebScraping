```python
# Do this before starts
# !pip install selenium
# !apt-get update
# !apt-get install chromium-chromedriver


from bs4 import BeautifulSoup as bs
import requests
import csv
from time import sleep
from selenium import webdriver

my_list = []
def main():
  
  #Chrome_Driver
  userAgent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('user-agent= {}'.format(userAgent))
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  driver = webdriver.Chrome('chromedriver', options = options)

  url='https://www.teamusa.org/usa-luge/official-clubs'

  driver.get(url)
  selenium_get = driver.page_source
  soup = bs(selenium_get, 'html.parser')

  info = soup.find('div', class_ = 'col-xs-12').find_all('p')
  
  info = info[2]
  info = info.text.split('\n\n')
  #print(info)
  #print(len(info))
  
  
  
  for a in info:
      my_dict={}

      b = a.split('\n')
      #print(b)
      #print(len(b))

      #Team
      if len(b) <= 7:
        my_dict['Name'] = b[0]
      else: 
        my_dict['Name'] = ''
      
      #Presidente do Clube
      if len(b) <= 7:
        my_dict['President'] = b[1]
      else:
        my_dict['President'] = ''

      #Address
      if len(b) <= 7:
        my_dict['Address'] = b[2]
      else:
        my_dict['Address'] = ''

      #City
      if len(b) <= 7:
        my_dict['City'] = b[3]
      else: 
        my_dict['City'] = ''
      
      #Phone Number
      if len(b) <= 7:
        my_dict['Phone'] = b[4]
      else: 
        my_dict['Phone'] = ''
      
      #Email
      if len(b) == 7:
        my_dict['Email'] = b[5]
      elif len(b) == 6:
        my_dict['Email'] = ''
      else:
        my_dict['Email'] = ''

      #Website
      if len(b) <= 7:
        my_dict['Website'] = b[-1]
      else:
        my_dict['Website'] = ''
      
      my_dict['Country'] = 'USA'
      my_dict['Affilated'] = 'USA Luge'
      my_dict['Recorded By'] = 'Eduardo Sávio'
      my_dict['Recorded Date'] = '10/02/2022'  
      my_list.append(my_dict) 
    
  return my_list

res = main()

keys = res[0].keys()

with open('USA_Luge.csv', 'w', newline = '', encoding = 'utf-8-sig') as csvfile:
  fieldnames = keys
  writer = csv.DictWriter(csvfile, fieldnames = fieldnames)

  writer.writeheader()
  writer.writerows(res)

csvfile.close()
```
