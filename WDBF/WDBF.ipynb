```python
from bs4 import BeautifulSoup as bs
import requests
import csv
from selenium import webdriver
import re

def find_name(soup):
  try:
      name = soup.find('p').text
  except:
      name=''
  
  return name


def find_email(soup):
  try:
     email = soup.find('a')['href'].replace('mailto:','')
  except:
      email=''
  
  return email

def find_country(soup):
  try:
      country_member = soup.find('h4', class_ = 'et_pb_module_header').text
  except:
      country_member = ''
  return country_member

def find_acronym(upper):
    acronym=''
    for a in upper:
        #check uppercase characters
        if a.isupper():
            acronym += a
    return acronym

def find_contact(soup):
  try:
      contact = soup.find_all('p')[1].text.split(' Joined:')[0].replace('Contact: ','')
  except:
      contact=''
  
  return contact



def main():
  
  
  url = 'https://worlddodgeballfederation.com/about-us/member-nations/'
  agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('user-agent= {}'.format(agent))
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  driver = webdriver.Chrome('chromedriver', options = options)

  driver.get(url)
  selenium_get = driver.page_source
  soup = bs(selenium_get, 'html.parser')
  
  #find every div with certain caracteres on the class -> 'et_pb_column et_pb_column_1_3 '
  #in that way, it is getting every country separated into variable info
  info = soup.findAll('div', {"class": re.compile('et_pb_column et_pb_column_1_3 ')})
  
  # here I'm iterating every country and getting every necessary info
  for a in info: 
    my_dict = {}
    
    my_dict['Name'] = find_name(a)
    my_dict['Acronym'] = find_acronym(my_dict['Name'])
    my_dict['Email'] = find_email(a)
    my_dict['Country'] = find_country(a)
    my_dict['Contact'] = find_contact(a)
    

    
    my_dict['Sport'] = 'Dodgeball'
    my_dict['Reach'] = 'National'
    my_dict['Category'] = 'Recognised'
    my_dict['Affilated'] = 'World Dodgeball Federation'
    my_dict['Recorded By'] = 'Eduardo SÃ¡vio'
    my_dict['Recorded Date'] = '17/02/2022'
    
    my_list.append(my_dict)
    
  return my_list
    
    #print(my_list)

my_list = []
res = main()

#save as xlsx
import pandas as pd
df = pd.DataFrame(data=res)
df.to_excel('World Dodgeball Federation.xlsx')
```
